### 2. Explore the Dataset

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
# Identificar los valores faltantes en todas las columnas
print("Cantidad de valores faltantes por columna:")
print(df.isnull().sum())
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
# Filtrar solo las columnas que tienen valores faltantes
missing_cols = df.columns[df.isnull().any()]

# Crear un mapa de calor solo con esas columnas
plt.figure(figsize=(12, 6))
sns.heatmap(df[missing_cols].isnull(), cbar=False, cmap='magma')
plt.title("Mapa de calor de columnas con valores faltantes")
plt.show()

```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
# Contar cuántas filas tienen valores faltantes en la columna 'Employment'
missing_employment = df['Employment'].isnull().sum()

print(f"Número de filas con valores faltantes en la columna 'Employment': {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
# Identificar el valor más frecuente en la columna 'Employment'
most_frequent_employment = df['Employment'].mode()[0]

print(f"El valor más frecuente en la columna 'Employment' es: {most_frequent_employment}")
```
