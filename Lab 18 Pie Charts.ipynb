### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("survey-data.csv")

# Separar las bases de datos en listas individuales
db_series = df['DatabaseWantToWorkWith'].dropna().str.split(',')

# Aplanar la lista y contar ocurrencias
all_dbs = pd.Series([db.strip() for sublist in db_series for db in sublist])
top_5_dbs = all_dbs.value_counts().nlargest(5)

# Crear gráfico circular
plt.figure(figsize=(8, 6))
plt.pie(top_5_dbs, labels=top_5_dbs.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Bases de Datos que los Encuestados Quieren Usar')
plt.axis('equal')  # Para que el gráfico sea circular
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("survey-data.csv")

# Separar los roles en listas individuales
dev_series = df['DevType'].dropna().str.split(',')

# Aplanar la lista y contar ocurrencias
all_roles = pd.Series([role.strip() for sublist in dev_series for role in sublist])
top_5_roles = all_roles.value_counts().nlargest(5)

# Crear gráfico circular
plt.figure(figsize=(8, 6))
plt.pie(top_5_roles, labels=top_5_roles.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Roles de Desarrollador entre los Encuestados')
plt.axis('equal')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("survey-data.csv")

# Extraer y limpiar la columna de sistemas operativos
opsys_series = df['OpSysProfessional use'].dropna().str.strip()

# Contar ocurrencias
opsys_counts = opsys_series.value_counts()

# Seleccionar los más comunes (puedes ajustar el número si lo deseas)
top_opsys = opsys_counts.nlargest(5)

# Crear gráfico circular
plt.figure(figsize=(8, 6))
plt.pie(top_opsys, labels=top_opsys.index, autopct='%1.1f%%', startangle=140)
plt.title('Sistemas Operativos Usados Profesionalmente por los Encuestados')
plt.axis('equal')
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("survey-data.csv")

# Paso 1: Separar por comas
lang_lists = df['LanguageHaveWorkedWith'].dropna().str.split(',')

# Paso 2: Separar también por barras "/"
all_langs = []
for sublist in lang_lists:
    for lang in sublist:
        parts = lang.strip().split('/')
        all_langs.extend([p.strip() for p in parts])

# Paso 3: Contar ocurrencias
lang_counts = pd.Series(all_langs).value_counts()

# Paso 4: Visualizar los 5 más comunes
top_5 = lang_counts.head(5)
plt.figure(figsize=(8, 6))
plt.pie(top_5, labels=top_5.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Lenguajes Separados Individualmente')
plt.axis('equal')
plt.show()

```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("survey-data.csv")

# Paso 1: Separar por comas
tools_lists = df['NEWCollabToolsHaveWorkedWith'].dropna().str.split(',')

# Paso 2: Separar también por barras "/"
all_tools = []
for sublist in tools_lists:
    for tool in sublist:
        parts = tool.strip().split('/')
        all_tools.extend([p.strip() for p in parts])

# Paso 3: Contar ocurrencias
tool_counts = pd.Series(all_tools).value_counts()

# Paso 4: Visualizar los 5 más comunes
top_5_tools = tool_counts.head(5)
plt.figure(figsize=(8, 6))
plt.pie(top_5_tools, labels=top_5_tools.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Herramientas de Colaboración Usadas Profesionalmente')
plt.axis('equal')
plt.show()
```
